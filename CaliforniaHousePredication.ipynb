# Spliting the Dataframe 🫡

```python
# deleting AveOccup columns from table
#dataset = dataset.drop(['AveOccup'], axis=1)
dataset.head()
```

```python
## Independent and dependent features
X = dataset
y = df.target
y
```

```python
## spliting our dataframe into test-field and train-field
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)
```

```python
X_train
```

## Standardizing the dataset 😉


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# save for future predication
save = X_train[0]
print(save)
saveRes = y_train[0]
print(saveRes)
```

## Training Model 😊

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
regression.coef_
```

### Predication on Test Data 😒

---

```python
y_pred = regression.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

```python
plt.scatter(y_test, y_pred)
```

### Using  Cross validation :✌️

```python
from sklearn.model_selection import cross_val_score
mse = cross_val_score(regression, X_train, y_train, scoring='neg_mean_squared_error', cv=10)
```

```python
np.mean(mse)
```

```python
y_pred_crossval = regression.predict(X_test)
r2_score(y_test, y_pred_crossval)
```
